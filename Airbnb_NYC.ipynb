```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Preparing data

airbnb_nyc=pd.read_csv('listings_LA2.csv') #Reading the csv file and making a dataframe
airbnb_nyc=airbnb_nyc.drop(['scrape_date','host_since','price','cleaning_fee','extra_people'],axis=1) #Dropping columns 'scrape_date','host_since','price','cleaning_fee','extra_people' from dataframe
rows_df =airbnb_nyc.shape[0] #No. of rows in the dataframe airbnb_nyc

facebook=[] #Empty list to be filled later with 0's and 1's depending on condition
email=[]
phone=[]
google=[]
jumio=[]
government_id=[]
kitchen=[]
refrigerator=[]
tv=[]
wifi=[]
air_conditioning=[]
heating=[]
washer=[]
dryer=[]
free_street_parking = []
free_parking_on_premises = []

for i in range(rows_df): #Appending 1 in facebook,email,phone,google,jumio,government_id "lists" which will eventually go into columns for the particular record if 'facebook','email','phone','google','jumio','government_id' respectively is present in host_verifications column of dataframe else appending 0. Doing it for all records in the dataframe hence the 'for' loop.  
    if 'facebook' in str(airbnb_nyc.host_verifications.iloc[i]):
        facebook.append(1)
    else:
        facebook.append(0)
for i in range(rows_df):
    if 'email' in str(airbnb_nyc.host_verifications.iloc[i]):
        email.append(1)
    else:
        email.append(0)
for i in range(rows_df):
    if 'phone' in str(airbnb_nyc.host_verifications.iloc[i]):
        phone.append(1)
    else:
        phone.append(0)
for i in range(rows_df):
    if 'google' in str(airbnb_nyc.host_verifications.iloc[i]):
        google.append(1)
    else:
        google.append(0)
for i in range(rows_df):
    if 'jumio' in str(airbnb_nyc.host_verifications.iloc[i]):
        jumio.append(1)
    else:
        jumio.append(0)
for i in range(rows_df):
    if 'government_id' in str(airbnb_nyc.host_verifications.iloc[i]):
        government_id.append(1)
    else:
        government_id.append(0)

for i in range(rows_df):
    if 'Kitchen' in str(airbnb_nyc.amenities.iloc[i]):
        kitchen.append(1)
    else:
        kitchen.append(0)
for i in range(rows_df):
    if 'Refrigerator' in str(airbnb_nyc.amenities.iloc[i]):
        refrigerator.append(1)
    else:
        refrigerator.append(0)
for i in range(rows_df):
    if 'TV' in str(airbnb_nyc.amenities.iloc[i]):
        tv.append(1)
    else:
        tv.append(0)
for i in range(rows_df):
    if 'Wifi' in str(airbnb_nyc.amenities.iloc[i]):
        wifi.append(1)
    else:
        wifi.append(0)
for i in range(rows_df):
    if 'Air conditioning' in str(airbnb_nyc.amenities.iloc[i]):
        air_conditioning.append(1)
    else:
        air_conditioning.append(0)
for i in range(rows_df):
    if 'Heating' in str(airbnb_nyc.amenities.iloc[i]):
        heating.append(1)
    else:
        heating.append(0)
for i in range(rows_df):
    if 'Washer' in str(airbnb_nyc.amenities.iloc[i]):
        washer.append(1)
    else:
        washer.append(0)
for i in range(rows_df):
    if 'Dryer' in str(airbnb_nyc.amenities.iloc[i]):
        dryer.append(1)
    else:
        dryer.append(0)
for i in range(rows_df):
    if 'Free street parking' in str(airbnb_nyc.amenities.iloc[i]):
        free_street_parking.append(1)
    else:
        free_street_parking.append(0)
for i in range(rows_df):
    if 'Free parking on premises' in str(airbnb_nyc.amenities.iloc[i]):
        free_parking_on_premises.append(1)
    else:
        free_parking_on_premises.append(0)
        
airbnb_nyc['facebook']=facebook #Creating column in dataframe and inserting values in the column through the list defined above
airbnb_nyc['email']=email
airbnb_nyc['phone']=phone
airbnb_nyc['google']=google
airbnb_nyc['jumio']=jumio
airbnb_nyc['government_id']=government_id
airbnb_nyc['kitchen']=kitchen
airbnb_nyc['refrigerator']=refrigerator
airbnb_nyc['tv']=tv
airbnb_nyc['wifi']=wifi
airbnb_nyc['air_conditioning']=air_conditioning
airbnb_nyc['heating']=heating
airbnb_nyc['washer']=washer
airbnb_nyc['dryer']=dryer
airbnb_nyc['free_street_parking'] = free_street_parking

print(airbnb_nyc)
```
